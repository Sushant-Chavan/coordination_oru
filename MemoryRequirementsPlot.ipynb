```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
```

```python
brsu_floor0_requirements = np.array([[4, 12, 20.5, 43.3],
                                     [6, 9.8, 10, 14.3],
                                     [70, 175.2, 345.5, 756.8]])

agp_basement_requirements = np.array([[5, 11.8, 22.6, 45.5],
                                      [5.3, 6.5, 8.3, 10.2],
                                      [102.8, 209, 464.9, 811.8]])

agp_floor4_requirements = np.array([[3.9, 12.2, 20, 38],
                                    [3.6, 6.5, 7.7, 10.8],
                                    [27, 63.9, 118.4, 214.4]])
```

```python
f = plt.figure(figsize=(15, 7.5))

plt.plot(brsu_floor0_requirements[2], label='EGraphs in BRSU_Floor0', linestyle='--', color='r')
plt.plot(agp_basement_requirements[2], label='EGraphs in AGP_Basement', linestyle='--', color='g')
plt.plot(agp_floor4_requirements[2], label='EGraphs in AGP_Floor4', linestyle='--', color='b')

plt.plot(brsu_floor0_requirements[0], label="Lightning in BRSU_Floor0", linestyle='-.', color='r')
plt.plot(agp_basement_requirements[0], label="Lightning in AGP_Basement", linestyle='-.', color='g')
plt.plot(agp_floor4_requirements[0], label="Lightning in AGP_Floor4", linestyle='-.', color='b')

plt.plot(brsu_floor0_requirements[1], label="Thunder in BRSU_Floor0", linestyle='-', color='r')
plt.plot(agp_basement_requirements[1], label="Thunder in AGP_Basement", linestyle='-', color='g')
plt.plot(agp_floor4_requirements[1], label="Thunder in AGP_Floor4", linestyle='-', color='b')

plt.title('Sizes of experience databases', fontsize=20)
plt.xticks(ticks=np.arange(4), labels=[10, 25, 50, 100])
plt.yticks(np.arange(0, 850, 50))
plt.xlabel("Number of experiences", fontsize=15)
plt.ylabel("Database size in KB's", fontsize=15)
plt.legend()
plt.grid()
plt.savefig("/home/suvich15/Sushant/MAS/MAS_RnD_Project/project-report/images/Results/Experiment3/Memory_AllPlanners.png")
plt.show()
```
